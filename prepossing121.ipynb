```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder

data = fetch_openml('car', as_frame=True).frame
data
```

```python
columns_to_encode = ['lug_boot','safety']

encoder = OrdinalEncoder(
    categories=[
        ['small','med','big'],
        ['low','med','high']
    ]
)

data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
data
```

```python
data = fetch_openml('adult',as_frame=True).frame
data
```

```python
data.occupation.value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_values = encoder.fit_transform(data[['occupation']])
new_colms = encoder.get_feature_names_out(['occupation'])
```

```python
encoded_values
```

```python
new_colms
```

```python
df_encoded = pd.DataFrame(encoded_values,columns=new_colms, index=data.index)

data_final =pd.concat(
    [data.drop(columns=['occupation']), df_encoded],
    axis=1
)

data_final
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,accuracy_score,precision_score,recall_score,f1_score

X,y = load_breast_cancer(return_X_y=True)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
model = KNeighborsClassifier()
model.fit(X_train_scaled,y_train)
```

```python
model.score(X_test_scaled, y_test)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,accuracy_score,precision_score,recall_score,f1_score

X,y = load_breast_cancer(return_X_y=True)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

model = KNeighborsClassifier()
model.fit(X_train,y_train)

model.score(X_test, y_test)
```
