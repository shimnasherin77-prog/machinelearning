```python
import tensorflow as tf
import numpy as np
print(tf.version)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)), #input layer
    tf.keras.layers.Dense(1)    #neuron layer
])
```

```python
#compile the model
model.compile(optimizer='sgd',loss='mean_squared_error')

#train the model
model.fit(x,y, epochs=200,verbose=0)

#test
print(model.predict(np.array([7])))

```

```python
# ⿢ Sleep Hours vs Energy Level
# Training Data:
# Hours Slept: [4, 5, 6, 7, 8]
# Energy Level (0-10): [2, 4, 5, 7, 8]
# Question: Predict energy level for 9 hours of sleep
```

```python
x = np.array([4,5,6,7,8])
y = np.array([2,4,5,7,8])

model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(1)    
])
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error')

model.fit(x,y, epochs=200,verbose=0)

print(model.predict(np.array([9])))
```

```python

# ⿣ Years of Experience vs Salary
# Training Data:
# Experience (years): [1, 2, 3, 4, 5]
# Salary (in $1000): [30, 35, 40, 45, 50]
# Question: Predict salary for 6 years of experience
```

```python
x = np.array([1,2,3,4,5])
y = np.array([30,35,40,45,50])

model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(1)    
])
```

```python
model.compile(optimizer='sgd',loss='mean_squared_error')

model.fit(x,y, epochs=200,verbose=0)

print(model.predict(np.array([6])))
```

```python
# ⿤ Temperature vs Ice Cream Sales
# Training Data:
# Temperature (°C): [20, 22, 24, 26, 28]
# Ice Cream Sales: [200, 250, 300, 350, 400]
# Question: Predict ice cream sales at 30°C
```

```python
x = np.array([20,22,24,26,28])
y = np.array([200,250,300,350,400])

model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(1)
])
```
