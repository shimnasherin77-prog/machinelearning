```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = load_iris()
X = data.data # feature only

#train k-means model
KMeans = KMeans(n_clusters=3,random_state=42)
KMeans.fit(X)

#get cluster labels
y_pred = KMeans.labels_
print("Cluster labels:",y_pred)

```

```python
import numpy as np
sample = np.array([[5.2, 3.8, 1.6, 0.2]])

sample_cluster = KMeans.predict(sample)
print("sample belongs to cluster:",sample_cluster)

plt.scatter(X[:, 0],X[:, 1], c=y_pred)
plt.scatter(sample[:, 0],sample[:, 1], s=200, marker='*')
plt.xlabel("Feature 1(Sepal Length)")
plt.ylabel("Feature 2(Sepal Width)")
plt.title("K-Means- new sample predicition")
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data[:,:2]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

n_clusters = 3

KMeans = KMeans(n_clusters=n_clusters,random_state=42)
KMeans.fit(X_scaled)

labels = KMeans.labels_
centroids = KMeans.cluster_centers_

plt.scatter(X_scaled[:,0],X_scaled[:,1],c=labels,cmap='viridis',edgecolors='k',s=50)
plt.scatter(centroids[:,0],centroids[:,1],c='red',marker='X',s=200,label='Centroids')
plt.title("K-Means clustering of Iris dataset")
plt.xlabel("Sepal Length(standardized)")
plt.ylabel("Sepal Width(standardized)")
plt.legend()
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("loandata.csv")
print(df)
```

```python
df = df.dropna()
```

```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = df[['employ','income']]

kmeans_model = KMeans(n_clusters=2, random_state=42)
kmeans_model.fit(X)

y_pred = kmeans_model.labels_
print("Cluster labels:", y_pred)

plt.figure(figsize=(10, 6))
plt.scatter(X['employ'], X['income'], c=y_pred, s=50, cmap='viridis')
plt.xlabel("Employment")
plt.ylabel("Income")
plt.title("K-Means Clustering of Customers")
plt.show()
```

```python
import numpy as np
sample = np.array([[14, 65]]) 
sample_cluster = kmeans_model.predict(sample)
print("Sample belongs to cluster:", sample_cluster)

plt.figure(figsize=(10, 6))
plt.scatter(X['employ'], X['income'], c=y_pred, s=100, cmap='viridis', label='Cluster')
plt.scatter(sample[:, 0], sample[:, 1], s=200, marker='*', c='red', label='New Sample')
plt.xlabel("Employment")
plt.ylabel("Income")
plt.title("K-Means Clustering with New Sample Prediction")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data
y = iris.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#apply pca to reduce dimensions to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
for i in range(len(set(y))):
    plt.scatter(X_pca[y == i,0],X_pca[y == i,1],label=f'class {i}', edgecolor='k', s=50)
plt.title('PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("customers.csv")
print(df)
```

```python
#Create a dataset using the given customer values (Annual Income and Spending Score).

#Tasks:
#1. Standardize the features.
#2. Apply K-Means clustering with 5 clusters.
#3. Plot a graph showing clusters using a scatter plot.
#4. Identify which type of customer group each cluster represents (for example: high income - high spending, low income - low spending, medium income - high spending,etc.).
```

```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = df[['AnnualIncome', 'SpendingScore']]

kmeans = KMeans(n_clusters=5, random_state=42)
y_kmeans = kmeans.fit_predict(X)

y_pred = kmeans.labels_
print("Cluster labels:", y_pred)

colors = ['red','blue','green','cyan','magenta']
plt.figure(figsize=(8, 6))
for i in range(5):
    plt.scatter(X.values[y_kmeans==i,0], X.values[y_kmeans==i,1],
                s=50, c=colors[i], label=f'Cluster {i}')
plt.title("K-Means Clustering of Customers")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

```python
import numpy as np
sample = np.array([[50, 60]]) 
sample_cluster = kmeans.predict(sample)
print("Sample belongs to cluster:", sample_cluster)

X_values = X.values

plt.figure(figsize=(10, 6))

plt.scatter(X_values[:, 0], X_values[:, 1], c=y_kmeans, s=100, cmap='viridis', label='Customers')

plt.scatter(sample[0, 0], sample[0, 1], s=200, marker='*', c='red', label='New Sample')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='yellow', marker='X', edgecolor='k', label='Centroids')

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("K-Means Clustering with New Sample Prediction")
plt.legend()
plt.show()
```
